```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import tkinter as tk
from tkinter import messagebox
from  sklearn.preprocessing import StandardScaler
```

```python
def training():
    # خواندن داده‌ها
    hepatitis_data = pd.read_csv('hepatitis_2_csv.csv')
    hepatitis = hepatitis_data.values

    X = hepatitis[:, :19]
    y = hepatitis[:, 19]


    sc=StandardScaler()
    X_s=sc.fit_transform(X)

    X_train, X_test, y_train, y_test = train_test_split(X_s, y, test_size=0.12, random_state=42)

    # ساخت مدل SVM
    svclassifier = SVC(kernel='linear')
    svclassifier.fit(X_train, y_train)
    print("Training complete")
    return svclassifier, X_test, y_test
```

```python
def testing(svclassifier, X_test, y_test):
    # پیش‌بینی با مدل
    predicted_class_name = svclassifier.predict(X_test)

    # محاسبه دقت
    accuracy = accuracy_score(y_test, predicted_class_name) * 100 #This was y_test, y_test and it should be y_test, predicted_class_name to get accuracy
    print(f"Accuracy: {accuracy:.2f}%")
```

```python
def main():
    # متغیرهای اولیه برای ذخیره مدل و داده‌ها
    svclassifier = None
    X_test = None
    y_test = None

    while True:
        # گرفتن ورودی از کاربر
        print("HEPATITIS DISEASE PREDICTION")
        print("1. Training")
        print("2. Testing")
        print("3. Close")

        choice = input("Select an option (1, 2, or 3): ")

        if choice == "1":
            # آموزش مدل
            svclassifier, X_test, y_test = training()
        elif choice == "2":
            print(f"svm_model before checking: {svclassifier}")
            if svclassifier is None:
               print("Please train the model first by selecting option 1.")
            else:
               testing(svclassifier, X_test, y_test)

        elif choice == "3":
            print("Closing program.")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
