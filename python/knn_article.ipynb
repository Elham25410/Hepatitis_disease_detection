```python
scaler = StandardScaler()
X_s = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_s, y, test_size=0.15, random_state=42)
```

```python
knn_model = KNeighborsClassifier(n_neighbors=7, metric='minkowski', p=2)
knn_model.fit(X_train, y_train)
```

```python
predicted_class_name = knn_model.predict(X_test)
```

```python
accuracy = (np.sum(predicted_class_name == y_test) / len(y_test)) * 100
print(f'Accuracy: {accuracy:.2f}%')
```
